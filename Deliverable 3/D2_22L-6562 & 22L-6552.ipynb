```python
# -----------------------------
# IMPORT REQUIRED LIBRARIES
# -----------------------------
import os
import cv2
import numpy as np
import re
from collections import Counter
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import seaborn as sns
from sklearn.utils.class_weight import compute_class_weight

# -----------------------------
# 1. HELPER: Extract angles from filename
# -----------------------------
def extract_angles(filename):
    match = re.match(r'.*?(-?\d+)P_(-?\d+)V_(-?\d+)H', filename)
    if match:
        pitch, vertical, horizontal = map(int, match.groups())
        return pitch, vertical, horizontal
    return None, None, None


# -----------------------------
# 2. HELPER: Categorize engagement based on angles
# -----------------------------
def categorize_gaze(pitch, vertical, horizontal):
    if abs(pitch) < 10 and abs(horizontal) < 10:
        return "attentive"
    elif abs(pitch) < 25 and abs(horizontal) < 25:
        return "distracted"
    else:
        return "disengaged"


# -----------------------------
# 3. NEW: Zoom + Center Crop (improves eye focus)
# -----------------------------
def zoom_and_center_crop(img, zoom_factor):
    h, w = img.shape[:2]

    new_h = int(h / zoom_factor)
    new_w = int(w / zoom_factor)

    start_h = (h - new_h) // 2
    start_w = (w - new_w) // 2
    end_h = start_h + new_h
    end_w = start_w + new_w

    cropped = img[start_h:end_h, start_w:end_w]
    cropped = cv2.resize(cropped, (224, 224))

    return cropped


# -----------------------------
# 4. Load dataset with improved preprocessing
# -----------------------------
def load_and_preprocess_images(data_path):
    images, labels = [], []

    for file in os.listdir(data_path):
        if file.endswith(".jpg"):
            img = cv2.imread(os.path.join(data_path, file))
            if img is None:
                continue

            # BETTER PREPROCESSING FOR ENGAGEMENT
            img = zoom_and_center_crop(img, zoom_factor=2)

            pitch, vertical, horizontal = extract_angles(file)

            if pitch is not None:
                category = categorize_gaze(pitch, vertical, horizontal)
                images.append(img)
                labels.append(category)

    return np.array(images), np.array(labels)


# -----------------------------
# 5. Optional visualization
# -----------------------------
def show_sample_images(X, y, category, num_samples=5):
    indices = [i for i, label in enumerate(y) if label == category]

    if len(indices) == 0:
        return

    plt.figure(figsize=(10, 2))
    for i, idx in enumerate(indices[:num_samples]):
        plt.subplot(1, num_samples, i + 1)
        plt.imshow(cv2.cvtColor(X[idx].astype(np.float32), cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.title(category)
    plt.show()


# -----------------------------
# 6. Load all datasets
# -----------------------------
X_list = []
y_list = []

for i in range(1, 6):
    if i >= 10:
        DATASET_PATH = f"/content/drive/MyDrive/dataset/00{i}/"
    else:
        DATASET_PATH = f"/content/drive/MyDrive/dataset/000{i}/"

    print("Loading and preprocessing images from", DATASET_PATH)

    tempX, tempy = load_and_preprocess_images(DATASET_PATH)

    tempX = tempX / 255.0  # Normalize

    print("Loaded:", len(tempX))
    print("Label distribution:", Counter(tempy))

    for category in ["attentive", "distracted", "disengaged"]:
        show_sample_images(tempX, tempy, category)

    X_list.append(tempX)
    y_list.append(tempy)

X = np.concatenate(X_list, axis=0)
y = np.concatenate(y_list, axis=0)

print("Final dataset size:", X.shape)
print("Final label distribution:", Counter(y))


# -----------------------------
# 7. Visualization of dataset distribution
# -----------------------------
label_counts = Counter(y)

plt.figure(figsize=(6, 4))
plt.bar(list(label_counts.keys()), list(label_counts.values()))
plt.title("Class Distribution")
plt.xlabel("Engagement Category")
plt.ylabel("Samples Count")
plt.show()


# -----------------------------
# 8. Encode labels + Train/Val split
# -----------------------------
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

le = LabelEncoder()
y_encoded = le.fit_transform(y)

X_train, X_val, y_train, y_val = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print("Train size:", X_train.shape)
print("Validation size:", X_val.shape)


# -----------------------------
# 9. Build Model: MobileNet
# -----------------------------
import tensorflow as tf
from tensorflow.keras.applications import MobileNet
from tensorflow.keras.layers import GlobalAveragePooling2D, Dropout, Dense
from tensorflow.keras.models import Model

base = MobileNet(
    include_top=False,
    weights='imagenet',
    input_shape=(224, 224, 3)
)

base.trainable = False

x = GlobalAveragePooling2D()(base.output)
x = Dropout(0.3)(x)
output = Dense(3, activation='softmax')(x)

model = Model(inputs=base.input, outputs=output)

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()


# -----------------------------
# 10. Training Stage 1 (head only)
# -----------------------------
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)

class_weights = dict(zip(np.unique(y_train), class_weights))
print("Class Weights:", class_weights)


# -----------------------------
# Training Stage 1 (head only)
# -----------------------------
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=10,
    batch_size=32,
    class_weight=class_weights
)


# -----------------------------
# Fine-Tuning Stage 2
# -----------------------------
base.trainable = True

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history_finetune = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=5,
    batch_size=32,
    class_weight=class_weights
)


# -----------------------------
# 12. Evaluation + Save
# -----------------------------
# Evaluate the model (keeps your original functionality)
loss, acc = model.evaluate(X_val, y_val)
print("Validation Accuracy:", acc)

# -------------------------------------------
# Get predictions for multiclass classification
# -------------------------------------------
y_pred_prob = model.predict(X_val)

# Convert probabilities to class labels (argmax)
y_pred = y_pred_prob.argmax(axis=1)

# If y_val is one-hot encoded, convert it back to class labels
if y_val.ndim > 1:
    y_true = y_val.argmax(axis=1)
else:
    y_true = y_val

# -------------------------------------------
# Import metrics
# -------------------------------------------
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# -------------------------------------------
# Calculate metrics (multiclass)
# -------------------------------------------
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='macro')
recall = recall_score(y_true, y_pred, average='macro')
f1 = f1_score(y_true, y_pred, average='macro')

print("Calculated Accuracy:", accuracy)
print("Precision (macro):", precision)
print("Recall (macro):", recall)
print("F1 Score (macro):", f1)

# -------------------------------------------
# Confusion Matrix
# -------------------------------------------
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()

# Optional: Full Classification Report
print("\nClassification Report:")
print(classification_report(y_true, y_pred))

# -------------------------------------------
# Save the model (keeps your original functionality)
# -------------------------------------------
model.save("/content/engagement_model.h5")
print("Model saved successfully!")
```
